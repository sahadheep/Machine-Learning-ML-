# Day 01 - Data Analysis with Pandas

## Objective
In this project, we will:

- Load a CSV dataset
- Inspect the dataset
- Handle missing values
- Remove duplicates
- Filter rows
- Create new columns
- Group data
- Merge DataFrames
- Calculate descriptive statistics
- Save the cleaned dataset


```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../datasets/raw_data.csv")
df.head()
```

```python
df.shape
df.columns
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df = df.drop("Cabin", axis=1)
df.shape
```

```python
df.columns
```

```python
df["Age"].mean()
```

```python
df["Age"] = df["Age"].fillna(df["Age"].mean())
```

```python
df["Age"].isnull().sum()
```

```python
df["Embarked"].mode()
```

```python
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
```

```python
df.isnull().sum()
```

```python
df["Embarked"].mode()
```

```python
df.duplicated().sum()
df.duplicated(subset = ["PassengerId"]).sum()
```

```python
df[df["Age"] > 30].shape
df[df["Age"]> 30].shape[0]
```

```python
df[df["Sex"] == "male"].shape[0]
```

```python
df[(df["Sex"] == "male") & (df["Age"] > 30)].shape[0]
```

```python
df.head()
```

```python
df["FamilySize"] = df["SibSp"]+df["Parch"]+1
df.shape
```

```python
df["IsAlone"] = (df["FamilySize"] == 1).astype(int)
df.shape
```

```python
df["AgeGroup"] = pd.cut(df["Age"] , bins = [0,12,18,35,60,100] , labels = ["Child","Teenager","Young Adult","Adult","Senior"])
```

#some group by operations 

```python
df.groupby("Sex")["Age"].mean()
```

```python
df.groupby("Sex")["Survived"].mean()
```

```python
df.groupby("Pclass")["Fare"].mean()
```

```python
df.groupby(["Sex" , "Pclass"])["Survived"].mean()
```

```python
df.corr(numeric_only=True)
```
